````python
# Re-load the CSV file after execution state reset
import pandas as pd

file_path = "/Users/raysun/Desktop/RAG-on-Tabular-Data/data_in_csv_format/test_100rows_in_csv.csv"
df = pd.read_csv(file_path)

# Define the prompt template
prompt_template = """
### Task  
Determine whether the given claim is **true or false** based on your knowledge and reasoning.  
This claim is related to the specified topic.

### Topic  
{topic}

### Claim  
"{claim}"

### Output Format  
- If the claim is **correct**, output: `1`  
- If the claim is **incorrect**, output: `0`  
- Do not provide explanations or additional text, just return `1` or `0`.  

### Example Outputs  
**Correct Claim**: `"The Eiffel Tower is located in Paris."` → `1`  
**Incorrect Claim**: `"The Amazon River is the longest river in the world."` → `0`
"""

# Generate prompts for each claim
df["prompt_without_table"] = df.apply(lambda row: prompt_template.format(topic=row["topic"], claim=row["claim"]), axis=1)
````

````python
# Define the prompt template with table data in JSON format
prompt_with_table_json_template = """
### Task  
Determine whether the given claim is **true or false** based on the provided structured table data.  
This claim is related to the specified topic.

### Topic  
{topic}

### Claim  
"{claim}"

### Table (JSON Format)  
```json
{table_json}

### Output Format  
- If the claim is **correct**, output: `1`  
- If the claim is **incorrect**, output: `0`  
- Do not provide explanations or additional text, just return `1` or `0`.  

### Example Outputs  
**Correct Claim**: `"The Eiffel Tower is located in Paris."` → `1`  
**Incorrect Claim**: `"The Amazon River is the longest river in the world."` → `0`
"""

# Generate prompts for each claim
df["prompt_with_table_json"] = df.apply(lambda row: prompt_with_table_json_template.format( topic=row["topic"], claim=row["claim"], table_json=row["table_JSON_format"]), axis=1)
````

````python
# Define the prompt template with table data in JSON format
prompt_with_table_json_template = """
### Task  
Determine whether the given claim is **true or false** based on the provided structured table data.  
This claim is related to the specified topic.

### Topic  
{topic}

### Claim  
"{claim}"

### Table (Markdown Format)  
{table_markdown}

### Output Format  
- If the claim is **correct**, output: `1`  
- If the claim is **incorrect**, output: `0`  
- Do not provide explanations or additional text, just return `1` or `0`.  

### Example Outputs  
**Correct Claim**: `"The Eiffel Tower is located in Paris."` → `1`  
**Incorrect Claim**: `"The Amazon River is the longest river in the world."` → `0`
"""

# Generate prompts for each claim
df["prompt_with_table_markdown"] = df.apply(lambda row: prompt_with_table_json_template.format( topic=row["topic"], claim=row["claim"], table_markdown=row["table_markdown_format"]), axis=1)
````
