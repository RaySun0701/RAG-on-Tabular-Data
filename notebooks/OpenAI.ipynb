```python
import pandas as pd
import openai
import time
```

```python
# Load the dataset with prompts
file_path = "/content/test_100rows_with_prompt.csv"
df = pd.read_csv(file_path)
```

```python
# Initialize OpenAI API client
client = openai.OpenAI(api_key="************")  # Replace with your actual API key
```

# Prompt **without** table

```python
# Function to call OpenAI GPT-4o API for each prompt
def get_gpt4o_output(prompt):
    try:
        response = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=[
                {"role": "system", "content": "You are a helpful assistant."},
                {"role": "user", "content": prompt},
            ],
            stream=False
        )
        # Extract and return the response
        return response.choices[0].message.content.strip()
    except Exception as e:
        return f"Error: {str(e)}"

# Ensure the column exists
if "prompt_without_table" not in df.columns:
    raise ValueError("Column 'prompt_without_table' not found in the dataset.")

# Call OpenAI GPT-4o for each claim and store the result
df["gpt4o_without_table"] = df["prompt_without_table"].apply(lambda prompt: get_gpt4o_output(prompt))
```

```python
df.head()
```

```python
df["gpt4o_without_table"] = df["gpt4o_without_table"].astype(str)
df["label"] = df["label"].astype(str)
```

```python
from sklearn.metrics import accuracy_score

# Get accuracy
accuracy = accuracy_score(df['label'], df['gpt4o_without_table'])
print(f"Accuracy: {accuracy:.4f}")
```

```python
true_counts = df["label"].value_counts()
print("True Label Counts:\n", true_counts)
```

```python
pred_counts = df["gpt4o_without_table"].value_counts()
print("\nPredicted Label Counts:\n", pred_counts)
```
